```python
# Abrir el archivo en modo de lectura
with open("rawText.txt", "r", encoding="utf-8") as file:
    # Leer todo el contenido del archivo
    content = file.read()

# Imprimir un fragmento del texto para verificar
print(content[:500])  # Muestra los primeros 500 caracteres
```

```python
import nltk
```

### NORMALIZACIÓN

```python
import string

# Texto original
texto = content

# Convertir a minúsculas
texto_normalizado = texto.lower()

# Eliminar puntuación
texto_normalizado = texto_normalizado.translate(str.maketrans('', '', string.punctuation))

# Mostrar un fragmento del texto normalizado
print(texto_normalizado[:500])  # Muestra los primeros 500 caracteres
```

### TOKENIZACIÓN

```python
from nltk.tokenize import word_tokenize, sent_tokenize
import nltk

# Descargar recursos necesarios
nltk.download('punkt')

# Tokenización por palabras
tokens_palabras = word_tokenize(texto_normalizado)

# Tokenización por oraciones
tokens_oraciones = sent_tokenize(texto_normalizado)

print("Tokenización por palabras:")
print(tokens_palabras)

print("\nTokenización por oraciones:")
print(tokens_oraciones)
```

## STOP WORDS

```python
from nltk.corpus import stopwords
import nltk

# Descargar stop words para español
nltk.download('stopwords')

# Cargar las stop words en español
stop_words = set(stopwords.words('spanish'))

# Eliminar stop words de los tokens
tokens_sin_stopwords = [word for word in tokens_palabras if word.lower() not in stop_words]

print("Tokens sin stop words:")
print(tokens_sin_stopwords)
```

## STEMMING

```python
from nltk.stem import SnowballStemmer

# Inicializar el stemmer para español
stemmer = SnowballStemmer('spanish')

# Aplicar stemming a los tokens sin stop words
tokens_stemmizados = [stemmer.stem(word) for word in tokens_sin_stopwords]

print("Tokens con stemming:")
print(tokens_stemmizados)
```

```python
print(len(tokens_stemmizados))
print(len(tokens_palabras))
```

```python
from collections import Counter

# Contar la frecuencia de cada token
conteo_tokens = Counter(tokens_stemmizados)

# Mostrar los 10 tokens más frecuentes
tokens_mas_repetidos = conteo_tokens.most_common(20)

print("Tokens más repetidos:")
for token, frecuencia in tokens_mas_repetidos:
    print(f"{token}: {frecuencia}")

print()
print(len(conteo_tokens))
```

```python
from collections import Counter

# Ejemplo: Conteo de tokens
conteo_tokens = Counter(tokens_stemmizados)

# Calcular el total de ocurrencias (frecuencia total)
total_ocurrencias = sum(conteo_tokens.values())

# Ordenar los tokens por frecuencia en orden descendente
tokens_ordenados = conteo_tokens.most_common()

# Calcular el umbral del 80% del total de ocurrencias
umbral_pareto = 0.8 * total_ocurrencias

# Seleccionar el 20% de los tokens que representan el 80% de las ocurrencias
acumulado = 0
tokens_pareto = []
for token, frecuencia in tokens_ordenados:
    if acumulado >= umbral_pareto:
        break
    tokens_pareto.append((token, frecuencia))
    acumulado += frecuencia

# Mostrar los resultados
print("Tokens seleccionados por la Regla de Pareto:")
for token, frecuencia in tokens_pareto:
    print(f"{token}: {frecuencia}")
```

### Retriever

```python
from langchain_aws.retrievers.bedrock import AmazonKnowledgeBasesRetriever
```

```python
kb_context_id = "BC3JS4SRQG"
kb_result_id = "MQPBML7IQ2"

# Base knowledge
retriever_context = AmazonKnowledgeBasesRetriever(
    knowledge_base_id = kb_context_id,
    retrieval_config = {
        "vectorSearchConfiguration": {
            "numberOfResults": 4,
            'overrideSearchType': "SEMANTIC" # optional
        }
    },
)

retriever_result = AmazonKnowledgeBasesRetriever(
    knowledge_base_id = kb_result_id,
    retrieval_config = {
        "vectorSearchConfiguration": {
            "numberOfResults": 4,
            'overrideSearchType': "SEMANTIC" # optional
        }
    },
)
```

```python
def get_retriver_docs(docs):    
    context = "\n".join([doc.page_content for doc in docs])  # Iteramos de forma síncrona si 'docs' es una lista
    return context
```

```python
async def get_retriver_context(params):
    docs = await retriever_context.ainvoke(params)
    return get_retriver_docs(docs)
```

```python
async def get_retriver_result(params):
    docs = await retriever_result.ainvoke(params)
    return get_retriver_docs(docs)
```

### Prompts

```python
from langchain.prompts import PromptTemplate
```

```python
def get_relevant_info():
    with open("1_get_relevant_info.sty", "r", encoding="utf-8") as file:
        prompt_content = file.read()
    
    return PromptTemplate(
        input_variables=["class_name"],
        template=prompt_content,
    )

```

```python
def get_exam_prompt():
    with open("6_get_exam_prompt.sty", "r", encoding="utf-8") as file:
        prompt_content = file.read()
    
    return PromptTemplate(
            input_variables=["prompt"],
            template=prompt_content
        )
```

### Definir el LLM


```python
from langchain_aws import BedrockLLM
from langchain_aws import ChatBedrock, ChatBedrockConverse

bedrock_model = "anthropic.claude-3-sonnet-20240229-v1:0"

llm_instance = ChatBedrockConverse(
    model=bedrock_model,
    temperature=0.02,
    max_tokens=None
)
```

### Parser y Filtro de Relevancia

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
```

```python
relevance = get_relevant_info() | llm_instance | parser
relevance_params = {
    "class_name": params["class_name"]
}
relevante = relevance.invoke(relevance_params)
```

### Proceso Parte 1 :  Resultados de Aprendizaje

```python
if relevante == 'NO':
  print("El tema que estás buscando no corresponde a un tema formativo, o actualmente no contamos con suficiente información para generar una clase sobre este asunto.")

else:
  prompt_template = get_base_prompt()
  classes = params["class_name"]
  context_query = f"{params['class_name']}, para grados {params['grade']}"
  context = get_retriver_context(context_query)

  dba_query = f"De los derechos basicos de aprendizaje dame la metodologia de enseñanza para la tematica de ({classes}) dirigida a estudiantes del grado {params['grade']}"
  dba = get_retriver_result(dba_query)

  learning_result_query = f"Dame información de los Estándares Básicos de Competencias y Orientaciones pedagógicas para la clase ({classes}) dirigida a estudiantes del grado {params['grade']}"
  learning_result = get_retriver_result(learning_result_query)

  chain = prompt_template | llm_instance | parser

  new_params = {
            "class_name": params["class_name"],
                "target_audience_edu_level":params["target_audience_edu_level"],
                "grade": params["grade"],
                "context": context,
                "dba": dba,
                "learning_result": learning_result,
                "date":  date.today()
            } 
  
  result_1 = chain.invoke(new_params)

  if params["target_audience_edu_level"] == "Preescolar":
      preschool_able = get_preschool_able_prompt() | llm_instance | parser
      preschool_able_params = {"result": result_1}
      result_1 = preschool_able.invoke(preschool_able_params)
```

### Proceso parte 2 : Conceptos clave

```python
key_template = get_key_concepts()

chain_key_concepts = key_template | llm_instance | parser
key_params = {
    "result_1": result_1,
    "class_name": params["class_name"],
    "target_audience_edu_level":params["target_audience_edu_level"],
    "grade": params["grade"],
    "context": context
}

result_2 = chain_key_concepts.invoke(key_params)

if params["target_audience_edu_level"] == "Preescolar":
    result_2 = preschool_able.invoke({"result": result_2})
```

### Proceso parte 3 : Actividades de aprendizaje

```python
result_1_2 = result_1 + "\n\n" + result_2
activities_template = get_activities_prompt()
chain_activities = activities_template | llm_instance | parser

activities_params = {
    "result_1_2": result_1_2,
    "class_name": params["class_name"],
    "target_audience_edu_level": params["target_audience_edu_level"],
    "grade": params["grade"],
    "context": context,
    "dba": dba
}

result_3 = chain_activities.invoke(activities_params)

if params["target_audience_edu_level"] == "Preescolar":
    result_3 = preschool_able.invoke({"result": result_3})
```

### Proceso parte 4 : Examen

```python
exam_template = get_exam_prompt()
chain_exam = exam_template | llm_instance | parser
examn_params = {
    "prompt": result_1 + '\n\n' + result_2 + '\n\n' + result_3 
}

result_4 = chain_exam.invoke(examn_params)

if params["target_audience_edu_level"] == "Preescolar":
    result_4 = preschool_able.invoke({"result": result_4})
```

# ¡Cocinemos Juntos!

Fecha de creación: 2025-01-23
Nivel educativo: Preescolar

## Resultados de aprendizaje:

1. Conocer los alimentos que comemos y cómo se ven.
2. Practicar nuestras manitas para cocinar.
3. Aprender a comer comidas ricas y saludables.
4. Jugar y trabajar juntos mientras cocinamos.

## Conceptos clave:

- **Alimentos**: Cosas que comemos para estar fuertes y sanos.
- **Cocinar**: Preparar comida calentándola o mezclándola.
- **Ingredientes**: Cosas que usamos para hacer una comida.
- **Nutrientes**: Cosas buenas en la comida que nos ayudan a crecer.
- **Receta**: Instrucciones para hacer una comida.
- **Saludable**: Algo que nos mantiene sanos y felices.
- **Trabajo en equipo**: Jugar y hacer cosas juntos con amigos.

## Metodologías

- **Aprender jugando**: Los niños aprenderán sobre los alimentos a través de juegos divertidos, canciones y actividades manuales.
- **Explorar con los sentidos**: Los niños tocarán, olerán y probarán diferentes alimentos para descubrir sus texturas, olores, colores y sabores.
- **Trabajar juntos**: Los niños jugarán y harán actividades en equipo, aprendiendo a compartir y ayudarse unos a otros.

## Actividades de aprendizaje

1. **Adivina el alimento**: Mostrar alimentos y pedir a los niños que adivinen su nombre, color, forma y textura.
2. **Explorar con las manos**: Dejar que los niños toquen, huelan y prueben ingredientes seguros.
3. **Cuentos y canciones**: Leer cuentos o cantar canciones divertidas sobre cocina y alimentos.
4. **Cocinar juntos**: El instructor mostrará cómo preparar recetas sencillas y saludables, explicando cada paso.
5. **Preparar recetas**: Los niños ayudarán a preparar recetas sencillas, siguiendo instrucciones y practicando habilidades manuales.

## Actividades prácticas

- **Ensalada de frutas**: Los niños cortarán (con herramientas seguras) y mezclarán frutas para hacer una ensalada de frutas colorida.
- **Pintar con alimentos**: Usar zanahorias, remolacha o espinacas para hacer pinturas naturales y dibujar.
- **Clasificar alimentos**: Separar alimentos por color, forma, textura o grupo alimenticio.

## Actividades reflexivas

- **Compartir experiencias**: Los niños contarán lo que más les gustó de los alimentos y actividades.
- **Hábitos saludables**: Hablar sobre la importancia de comer alimentos nutritivos y variados para estar sanos y fuertes.
- **Trabajar en equipo**: Hablar sobre cómo se ayudaron unos a otros durante las actividades de cocina.
